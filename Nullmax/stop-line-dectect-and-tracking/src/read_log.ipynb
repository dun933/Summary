```python
# coding=utf-8
# 该文件用来提取训练log，去除不可解析的log后使log文件格式化，生成新的log文件供可视化工具绘图
import inspect
import os
import random
import sys

import re
def extract_log(log_file,new_log_file,key_word):
    i = 0
    with open(log_file, 'r') as f: 
        with open(new_log_file, 'w') as train_log:   
            for line in f.readlines():
                tmp = re.findall(key_word + ',\d+,\d+,\d+.\d+',line)
                if len(tmp)==0:
                    break
                st = tmp[0]
                batch = re.search('\d+',st).group()
                loss = re.search('30000,\S+',st).group()[5:]
                print(batch,loss)
                i = i+1
                train_log.write(batch+loss+'\n')
    f.close()
    train_log.close()
     
extract_log('iou/log.csv','iou_loss.txt','BATCH')
```
