```python
import os
import random

trainval_percent = 0.9# trainval数据集占所有数据的比例
train_percent = 0.9# train数据集占trainval数据的比例
filepath = '/home/user/ocr/yinhang_ssd/data/YH/VOC2007/JPEGImages'
txtsavepath = '/home/user/ocr/yinhang_ssd/data/YH/VOC2007/ImageSets/Main/'
total_txt = os.listdir(filepath)

num=len(total_txt)
list=range(num)
tv=int(num*trainval_percent)
tr=int(tv*train_percent)
trainval= random.sample(list,tv)
train=random.sample(trainval,tr)

ftrainval = open(txtsavepath + 'trainval.txt', 'w')
ftest = open(txtsavepath +'test.txt', 'w')
ftrain = open(txtsavepath + 'train.txt', 'w')
fval = open(txtsavepath + 'val.txt', 'w')

for i in list:
    print(total_txt[i])
    name=total_txt[i][:-4]+'\n'
    if i in trainval:
        ftrainval.write(name)
        if i in train:
            ftrain.write(name)
        else:
            fval.write(name)
    else:
        ftest.write(name)

ftrainval.close()
ftrain.close()
fval.close()
ftest.close()
```
